IMPORTING REQUESTS AND BEAUTIFULSOUP LIBRARY

```python
import requests
from bs4 import BeautifulSoup
```

Function to Get and Parse HTML content from a Wikipedia page

```python
def get_wikipedia_html(url):
    page = requests.get(url)
    if page.status_code == 200:
        return BeautifulSoup(page.content, 'html.parser')
    else:
        return None
```

Function to Extract the Article Title

```python
def extract_title(soup):
    if soup:
        return soup.title.string
    return None
```

Function to Extract Article Text with Headings

```python
def extract_article_text(soup):
    if soup is None:
        return None

    content = {}
    current_heading = ""

    for tag in soup.find_all(['h2', 'h3', 'h4','h5', 'h6', 'p']):
        if tag.name in ['h2', 'h3', 'h4', 'h5', 'h6']:
            current_heading = tag.get_text(strip=True)
            content[current_heading] = []
        elif tag.name == 'p':
            if current_heading:
                content[current_heading].append(tag.get_text(strip=True))

    return content
```

Wrap up to a Single function

```python
import requests
from bs4 import BeautifulSoup

def parse_wikipedia_page(url):

    response = requests.get(url)
    if response.status_code != 200:
        return None

    soup = BeautifulSoup(response.content, 'html.parser')

    title = soup.title.string if soup.title else 'No title found'

    content = {}
    current_heading = ""
    for tag in soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6', 'p']):
        if tag.name in ['h1', 'h2', 'h3', 'h4', 'h5', 'h6']:
            current_heading = tag.get_text(strip=True)
            content[current_heading] = []
        elif tag.name == 'p':
            if current_heading:
                content[current_heading].append(tag.get_text(strip=True))


    links = set()
    for link in soup.find_all('a', href=True):
        href = link['href']
        if href.startswith('/wiki/') and ':' not in href:
            full_url = f"https://en.wikipedia.org{href}"
            links.add(full_url)

    return {
        'title': title,
        'content': content,
        'links': list(links)
    }
```

```python
if __name__ == "__main__":
    # Get the URL from the user
    url = input("Enter a Wikipedia URL: ")
    data = parse_wikipedia_page(url)


    if data:
        print(f"Title: {data['title']}")

        print("\nHeadings with Content:")
        for heading, paragraphs in data['content'].items():
            print(f"Heading: {heading}")
            for paragraph in paragraphs:
                print(f" - {paragraph}")

        print("\nLinks to other Wikipedia pages:")
        for link in data['links']:
            print(link)
    else:
        print("Failed to retrieve the Wikipedia page.")
```
