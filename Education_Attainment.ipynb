```python
# import necessary libaries
from ydata_profiling import ProfileReport
import pandas as pd
```

```python
#read the education_attainment dataset
Edu_A_df = pd.read_csv("EDUCATION_ATTAINMENT.csv")
```

```python
# Generation information of the dataset
Edu_A_df.info()
```

```python
# Descriptive statistics summary
Edu_A_df.describe()
```

```python
# Creating a Profile for the dataset
profile = ProfileReport(Edu_A_df, title = "Pandas Profile Report", explorative=True)
```

```python
# Converting the profile to an html file
profile.to_file(output_file='ProfileReport.html')
```
