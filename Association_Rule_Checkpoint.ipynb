```python
toy_dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

            ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

            ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

            ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

              ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
import mlxtend
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd
```

```python
#Apply one-hot-encoding on our dataset

te = TransactionEncoder()
te_array = te.fit(toy_dataset).transform(toy_dataset)
```

```python
# Creating a DataFrame from our Numpy array
df = pd.DataFrame(te_array, columns = te.columns_)
df
```

```python
# Applying minimum support of 60% to know how frequent customers bought each item or which item is mostly bought by customers
from mlxtend.frequent_patterns import apriori
frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)
frequent_itemsets
```

```python
#From the support count, we see that the following items : Hat, Pants, Scarf, Skirt which include are mostly bought by customers from the troy dataset
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7) # associate items
```

```python
# From the Confidence Association Rule Result
# This implies that 75% of the transactions containing Scarf have Pants
# this means that we have 75% chance that customers would buy Pants after buying Scarf
# Similarly, there is 100% chance that customers would buy Pants after buying Skirt
```

```python
# Applying Association Rule based on Lift

from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets,metric="lift",min_threshold=1.25)
```

```python
#using minimum threshold of 1.25 implies that 60 % of customers who bought skirt are likely to buy pants
#while 80% of customers who bought pants are likely to buy skirts. Hence, customers who bought skirt also bought pants
```

```python
market_basket = pd.read_csv("Market_Basket_Optimisation.csv")
market_basket.head()
```

```python
market_basket.info()
```

```python
# Getting the list of transactions from the dataset
transactions = []
for i in range(0, len(market_basket)):
    transactions.append([str(market_basket.values[i,j]) for j in range(0, len(market_basket.columns))])
```

```python
print(transactions)
```

```python
from itertools import permutations

# Extract unique items.
flattened = [item for transaction in transactions for item in transaction]
items = list(set(flattened))
```

```python
if 'nan' in items: items.remove('nan')
print(list(items))
```

```python
# Compute and print rules.
rules = list(permutations(items, 2))
print('# of rules:',len(rules))
print(rules[:5])
```

```python
encoder = TransactionEncoder().fit(transactions)
ben_encode = encoder.transform(transactions)
ben_encode = pd.DataFrame(ben_encode, columns = encoder.columns_).drop('nan', axis=1)
ben_encode.head()
```

```python
import plotly.express as px
item_counts = ben_encode.sum().sort_values(ascending=False).reset_index()
item_counts.columns = ['Item', 'Count']

# Plot the bar graph using Plotly
fig = px.bar(item_counts, x='Item', y='Count', title='Item Frequency in Transactions')
fig.show()
```

```python
# Applying minimum support of 15% to know how frequent customers bought each item or which item is mostly bought by customers
from mlxtend.frequent_patterns import apriori
frequent = apriori(ben_encode, min_support=0.01, use_colnames=True).round(2)
frequent
```

```python
rules = association_rules(frequent, metric="confidence", min_threshold=0.1)

# Display support for item pairs
pairs_support = frequent[frequent['itemsets'].apply(lambda x: len(x) == 2)]
print(pairs_support)
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules(frequent,metric="confidence",min_threshold=0) # associate items
```
